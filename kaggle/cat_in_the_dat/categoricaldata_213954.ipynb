```python
## 데이터 둘러보기

train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cat-in-the-dat/train.csv')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cat-in-the-dat/test.csv')
submission = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cat-in-the-dat/sample_submission.csv')

### 데이터 합치기
all_data_temp = pd.concat([train, test], ignore_index=True)
print(all_data_temp)
```

```python
### bin_0 ~ bin_4 의 고유값 출력
for i in range(5):
    feature = 'bin_' + str(i)
    print(f'{feature} unique: {all_data[feature].unique()}')
```

```python
# 이진 피처 인코딩
# bin_0, bin_1, bin_2 피처는 0과 1
# bin_3은 T와 F, bin_4는 Y와 N
all_data['bin_3'] = all_data['bin_3'].map({'F':0, 'T':1})
all_data['bin_4'] = all_data['bin_4'].map({'N':0, 'Y':1})
```

```python
### ord_0 ~ ord_5 의 고유값 출력
for i in range(6):
    feature = 'ord_' + str(i)
    print(f'{feature} unique: {all_data[feature].unique()}')
```

```python
# 순서형 피처 인코딩
# ord_0 피처는 이미 숫자
# ord_1과 ord_2 인코딩
ord1dict = {'Novice':0, 'Contributor':1, 'Expert':2,
            'Master':3, 'Grandmaster':4}
ord2dict = {'Freezing':0, 'Cold':1, 'Warm':2, 'Hot':3,
            'Boiling Hot':4, 'Lava Hot':5}

all_data['ord_1'] = all_data['ord_1'].map(ord1dict)
all_data['ord_2'] = all_data['ord_2'].map(ord2dict)
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_345 = ['ord_3','ord_4','ord_5']

ord_encoder = OrdinalEncoder()

all_data[ord_345] = ord_encoder.fit_transform(all_data[ord_345])
```

```python
# 명목형 피처 인코딩
# 순서를 무시해도 무관하므로 원-핫 인코딩

nom_features = ['nom_'+str(i) for i in range(10)]

from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder()

encoded_nom_matrix = onehot_encoder.fit_transform(all_data[nom_features])

all_data = all_data.drop(nom_features, axis=1)
```

```python
# 날짜 피처 인코딩
date_features = ['day','month']

encoded_date_matrix = onehot_encoder.fit_transform(all_data[date_features])

all_data = all_data.drop(date_features, axis=1)
```

```python
### 피처 요약표 Feature Summarized Table
def resumetable(df):
    print(f'Shape of dataset: {df.shape}')
    summary = pd.DataFrame(df.dtypes, columns=['Data Type'])
    summary = summary.reset_index()
    summary = summary.rename(columns={'index': 'Feature'})

    summary['Missing'] = df.isnull().sum().values
    summary['Unique'] = df.nunique().values
    summary['First value'] = df.loc[0].values
    summary['Second value'] = df.loc[1].values
    summary['Third value'] = df.loc[2].values

    return summary

resumetable(train)
```

```python
## 베이스라인 모델
## 데이터 불러오기

train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cat-in-the-dat/train.csv', index_col='id')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cat-in-the-dat/test.csv', index_col='id')
submission = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cat-in-the-dat/sample_submission.csv', index_col='id')
```

```python
### 피처 엔지니어링
# 데이터 합치기
all_data = pd.concat([train, test])

all_data
```

```python
all_data = all_data.drop('target', axis=1)

all_data.head()
```

```python
all_data.shape
```

```python
### 원-핫 인코딩
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
all_data_encoded = encoder.fit_transform(all_data)

all_data_encoded
```

```python
print(all_data_encoded)
```

```python
### 데이터 나누기

num_train = len(train)

X_train = all_data_encoded[:num_train]
X_test = all_data_encoded[num_train:]

y = train['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X_train, y,
                                                     test_size=0.1,
                                                     stratify=y,
                                                     random_state=10)
```

```python
### 모델 훈련

from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(max_iter=1000, random_state=42)

logistic_model.fit(X_train, y_train)
```

```python
### 모델 성능 검증
# predict(): 타깃값 자체 0 혹은 1을 예측
# predict_proba() : 타깃값의 확률을 예측

logistic_model.predict_proba(X_valid)
```

```python
logistic_model.predict(X_valid)
```

```python
# 검증 데이터로 타깃 예측값을 구함
y_valid_preds = logistic_model.predict_proba(X_valid)[:,1]
```

```python
# ROC AUC 평가함수
from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(y_valid, y_valid_preds)

print(f'검증 데이터 ROC AUC : {roc_auc: .4f}')
```

```python
# 타깃값이 1일 확률 예측
y_preds = logistic_model.predict_proba(X_test)[:,1]
y_preds
```

```python
## 제출

submission['target'] = y_preds
submission.to_csv('/content/drive/MyDrive/Colab Notebooks/cat-in-the-dat/cat_submission.csv')
```

```python
import optuna
from sklearn.model_selection import StratifiedKFold

kf = StratifiedKFold(5)
```

```python
from sklearn.model_selection import cross_validate, cross_val_score

def objective(trial):
    clasas_weight=trial.suggest_uniform('clasas_weight', 1, 2)
    model=LogisticRegression(C=0.12, class_weight={0:1, 1:clasas_weight},
                             max_iter=10000, solver='lbfgs', n_jobs=-1)
    score=-cross_val_score(model, X_train, y_train, cv=kf, scoring='roc_auc').mean()
    return score
study=optuna.create_study()
```

```python
study.optimize(objective, n_trials=50)
```

```python
print(study.best_params)
print(-study.best_value)
params=study.best_params
```
